#### Eksekusi Query dengan Pandas

```python
# import library
import pandas as pd

# Ambil data dan batasi 17314 (data sampel pada bulan januari)
data = pd.read_excel("Coffee Shop Sales.xlsx")
data_sample = data.head(17314).copy()

# 1. Penjualan per Lokasi Toko
sales_per_location = data_sample.groupby(['store_id', 'store_location'])[['transaction_qty', 'unit_price']].apply(lambda x: (x['transaction_qty'] * x['unit_price']).sum()).reset_index(name='total_sales')

# 2. Penjualan per Kategori Produk
sales_per_category = data_sample.groupby('product_category')[['transaction_qty', 'unit_price']].apply(lambda x: (x['transaction_qty'] * x['unit_price']).sum()).reset_index(name='total_sales')

# 3. Penjualan per Produk
sales_per_product = data_sample.groupby(['product_id', 'product_detail'])[["transaction_qty", "unit_price"]].apply(lambda x: (x["transaction_qty"] * x["unit_price"]).sum()).reset_index(name='total_sales')

# 4. Penjualan berdasarkan Waktu Transaksi
sales_per_time = data_sample.groupby(['transaction_date', 'transaction_time'])[['transaction_qty', 'unit_price']].apply(lambda x: (x['transaction_qty'] * x['unit_price']).sum()).reset_index(name='total_sales')

# 5. Harga Unit dan Pengaruhnya terhadap Penjualan
sales_per_unit_price = data_sample.groupby('unit_price')['transaction_qty'].sum().reset_index(name='total_quantity')
sales_per_unit_price['total_sales'] = sales_per_unit_price['unit_price'] * sales_per_unit_price['total_quantity']

# 6. Distribusi Jumlah Transaksi dan Kuantitas Produk yang Terjual
transaction_distribution = data_sample.groupby('transaction_id')['transaction_qty'].sum().reset_index(name='total_quantity')

# Display the results
print(sales_per_location.head()) 
print(sales_per_category.head())
print(sales_per_product.head()) 
print(sales_per_time.head()) 
print(sales_per_unit_price.head()) 
print(transaction_distribution.head())
```

Penjelasan: 

- sales per lokasi toko = menunjukkan total penjualan untuk setiap toko berdasarkan lokasi. Toko di Astoria memiliki total penjualan sebesar 27313.66, diikuti oleh Hell's Kitchen dengan 27820.65, dan Lower Manhattan dengan 27826543.43.

- sales per kategori produk = menunjukkan total penjualan untuk setiap kategori produk. Bakery merupakan kategori dengan penjualan tertinggi sebesar 9535.75.

- sales per produk = menunjukkan total penjualan untuk setiap produk spesifik. Produk "Brazilian - Organic" memiliki total penjualan sebesar 360.00, dan seterusnya untuk produk lainnya.

- sales berdasar waktu transaksi = menunjukkan total penjualan berdasarkan waktu transaksi. Setiap baris mewakili transaksi tertentu dengan tanggal dan waktu spesifik serta total penjualan yang dihasilkan pada saat itu.

- pengaruh unit price terhadap total sales = menunjukkan bagaimana harga unit produk mempengaruhi total penjualan. Harga unit yang lebih rendah cenderung memiliki total penjualan lebih tinggi karena kuantitas yang lebih banyak terjual.

- distribusi jumlah transaksi dan kuantitas produk yang terjual = menunjukkan jumlah total kuantitas produk yang terjual dalam setiap transaksi.

#### **3. Statistik Inferensial**

**Masalah 1:** Analisis perbedaan penjualan antara kategori produk menggunakan ANOVA.


```python
import pandas as pd
unique_product_categories = data_sample['Product Category'].unique()
unique_product_categories
```

```python
# Import library
# Untuk analisis varian satu arah (ANOVA) terhadap beberapa sampel
from scipy.stats import f_oneway    
    
# Menghitung total penjualan per transaksi per kategori produk
sales_per_category_transaction = data_sample.groupby(['Transaction ID', 'Product Category']).apply(lambda x: (x['Transaction Quantity'] * x['Unit Price']).sum()).reset_index(name='Total Sales')

# Mengambil penjualan per kategori produk
coffee_sales = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Coffee']['Total Sales']
tea_sales = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Tea']['Total Sales']
bakery_sales = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Bakery']['Total Sales']
drinking_chocolate_sales = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Drinking Chocolate']['Total Sales']
flavours_sales  = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Flavours']['Total Sales']
loose_tea_sales  = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Loose Tea']['Total Sales']
coffee_beans_sales  = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Coffee beans']['Total Sales']
packaged_chocolate_sales  = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Packaged Chocolate']['Total Sales']
branded_sales  = sales_per_category_transaction[sales_per_category_transaction['Product Category'] == 'Branded']['Total Sales']


print(f'Average Penjualan Coffe: {coffee_sales.mean():.2f}')
print(f'Average Penjualan Tea: {tea_sales.mean():.2f}')
print(f'Average Penjualan Bakery: {bakery_sales.mean():.2f}')
print(f'Average Penjualan Drinking Chocolate: {drinking_chocolate_sales.mean():.2f}')
print(f'Average Penjualan Flavours: {flavours_sales.mean():.2f}')
print(f'Average Penjualan Loose Tea: {loose_tea_sales.mean():.2f}')
print(f'Average Penjualan Coffee beans: {coffee_beans_sales.mean():.2f}')
print(f'Average Penjualan Packaged Chocolate: {packaged_chocolate_sales.mean():.2f}')
print(f'Average Penjualan Branded: {branded_sales.mean():.2f}')
```

**Hipotesis:**
- H0: Tidak terdapat perbedaan penjualan yang signifikan pada keempat kategori produk.
- Ha: Terdapat perbedaan penjualan yang signifikan pada keempat kategori produk.

```python
# Melakukan uji ANOVA
anova_result = f_oneway(coffee_sales, tea_sales, bakery_sales, drinking_chocolate_sales)

# Menampilkan hasil
anova_result
```

**Insight:**

Data rata-rata penjualan untuk keempat kategori produk:
- Coffee: 4.60
- Tea: 4.30
- Bakery: 3.61
- Drinking Chocolate: 6.25

Interpretasi hasil:

- Jika p-value < 0.05: Terdapat perbedaan penjualan yang signifikan di antara kategori produk.
- Jika p-value  0.05: Tidak terdapat perbedaan penjualan yang signifikan di antara kategori produk.

Berdasarkan hasil ini, perusahaan dapat mempertimbangkan untuk fokus pada produk-produk dengan penjualan tertinggi atau mencari cara untuk meningkatkan penjualan produk dengan rata-rata lebih rendah.

Hasil ANOVA:
- F-statistic: 697.282

- p-value: 0.0

    P-value yang sangat kecil (0.0) menunjukkan bahwa ada perbedaan yang sangat signifikan antara rata-rata penjualan dari keempat kategori produk yang diuji. Secara umum, jika p-value kurang dari tingkat signifikansi (0.05), maka akan menolak hipotesis nol.

Karena ada perbedaan yang signifikan dalam penjualan antara kategori produk, penting untuk memahami faktor-faktor yang mempengaruhi penjualan di setiap kategori. Strategi pemasaran yang disesuaikan untuk setiap kategori dapat meningkatkan penjualan secara keseluruhan.

Hasil uji ANOVA menunjukkan bahwa Coffee beans dan Branded products memiliki rata-rata penjualan yang sangat tinggi. Investasi dalam pemasaran dan promosi untuk kategori ini dapat meningkatkan pendapatan secara substansial.
